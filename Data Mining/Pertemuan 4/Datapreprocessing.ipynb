An independent variable (x) is a variable that you purposely change in order to see the effect in the dependent variable(y).
For example, in equation y = mx + c, x is the independent variable and y is the dependent variable because you change/tune x to see its change on y. Isnâ€™t it?

**Check for missing values**

Understanding the ways to manage missing values is very important. As we discussed earlier, real-world data can often has missing values. If the missing values are not handled properly then the researcher may end up drawing incorrect reference about the data. And handling the missing values is important because many machine learning models do not support data with missing values.
We can always count the number of columns with missing values.

```python
dataset.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy="median")
X[:, 0:9] = imputer.fit_transform(X[:, 0:9])
```
