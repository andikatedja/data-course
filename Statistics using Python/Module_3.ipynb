# T-test

```python
ages = [10, 30, 20, 30, 20, 35, 35, 67, 87, 45, 34, 16, 55, 17, 16, 65, 19, 10, 43, 23, 67, 30, 34, 72, 81, 35, 67, 87, 45, 34, 16]
```

```python
len(ages)
```

```python
import numpy as np
```

```python
np.mean(ages)
```

```python
sample_size = 10
sample_age = np.random.choice(ages, sample_size)
sample_age
```

### 1 Sample T-test

```python
from scipy.stats import ttest_1samp
```

```python
ttest, p_value = ttest_1samp(sample_age, np.mean(ages))
```

```python
p_value
```

```python
if p_value < 0.05:
    print('We are rejecting our null hypotesis')
else:
    print('We are accepting our null hypotesis')
```

### 2 Sample T-test

```python
import numpy as np
import scipy.stats as stats
```

```python
classa_height = stats.poisson.rvs(mu=30, size=60)
classb_height = stats.poisson.rvs(mu=33, size=60)
```

```python
classb_height
```

```python
np.mean(classa_height)
```

```python
np.mean(classb_height)
```

```python
from scipy.stats import ttest_ind
```

```python
t_test, p_value = ttest_ind(a=classa_height, b=classb_height, equal_var=False)
```

```python
p_value
```

```python
if p_value < 0.05:
    print('We are rejecting our null hypotesis')
else:
    print('We are accepting our null hypotesis')
```

### Paired T-test

```python
import pandas as pd
```

```python
df = pd.read_csv('Datasets/blood_pressure.csv')
df
```

```python
df[['bp_before', 'bp_after']].describe()
```

#### Null hypotesis, H0 -> mu = 0
#### Alternative hypotesis, H1 -> mu /= 0 

```python
from scipy.stats import ttest_rel
```

```python
t_test, p_value = ttest_rel(df['bp_before'], df['bp_after'])
p_value
```

```python
if p_value < 0.05:
    print('We are rejecting our null hypotesis')
else:
    print('We are accepting our null hypotesis')
```
