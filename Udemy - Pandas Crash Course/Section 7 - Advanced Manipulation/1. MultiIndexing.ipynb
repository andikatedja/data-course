```python
import pandas as pd
```

```python
dfo = pd.read_csv('flights2.csv.gz', low_memory=False)
print(dfo.columns)
dfo.head()
```

```python
df = dfo.set_index(['DESTINATION_AIRPORT', 'AIRLINE'])
df = df.sort_index()
df.head()
```
