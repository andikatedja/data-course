```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

## Bar Plots

```python
chest_pain = df.groupby(by='cp').median().reset_index()
chest_pain.head()
```

```python
chest_pain.plot.bar(x='cp', y='age');
```

```python
chest_pain.plot.bar(x='cp');
```

```python
fig, ax = plt.subplots()
ax.bar(chest_pain['cp'], chest_pain['age'], label='age')
ax.set_xlabel('cp')
ax.set_xticks(chest_pain['cp'])
ax.legend();
```

```python
fig, ax = plt.subplots()
ax.bar(chest_pain['cp'], chest_pain['age'], label='age', 
edgecolor='k', color=['red', 'blue', 'green', 'black'])
ax.set_xlabel('cp')
ax.set_xticks(chest_pain['cp']);
```

```python
ax = sns.barplot(x='cp', y='age', errcolor='k', capsize=0.1, data=df)
```

## Scatter Plots

```python
df.plot.scatter('age', 'trestbps');
```

```python
fig, ax = plt.subplots()
ax.scatter(df['age'], df['trestbps'], marker='^',
s=60, c=df['age'], edgecolors='k', alpha=0.5)
ax.set_xlabel('age')
ax.set_ylabel('trestbps');
```

```python
sns.scatterplot(x='age', y='trestbps', hue='age', s=30, edgecolor='none', data=df)
```

## Line Plots

```python
ages = df.groupby('age').median().reset_index()
ages.head()
```

```python
ages.plot.line('age', 'chol');
```

```python
ages.plot.line('age', ['chol', 'trestbps']);
```

```python
fig, ax = plt.subplots()
ax.plot(ages['age'], ages['trestbps'], ls=':', lw=1.7)
ax.set_xlabel('age')
ax.set_ylabel('trestbps');
```

```python
sns.lineplot(data=ages, x='age', y='trestbps');
```
