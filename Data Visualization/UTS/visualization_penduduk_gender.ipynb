```python
# Membaca CSV Offline
table = pd.read_csv('penduduk_gender_head.csv')
table.head(15)
```

```python
# Cleansing Data
table = table.drop(table.iloc[:, 9:], axis=1)
```

```python
# Plot jumlah penduduk laki laki wni berdasarkan kelurahan
plt.figure(figsize=(15, 10))
x_label = table['NAMA KELURAHAN']
plt.bar(x=np.arange(len(x_label)),height=table['LAKI-LAKI WNI'])
plt.xticks(np.arange(len(x_label)), table['NAMA KELURAHAN'], rotation=90)
plt.xlabel('Kelurahan di Jakarta pusat')
plt.ylabel('Jumlah Penduduk Laki - Laki')
plt.title('Persebaran Jumlah Penduduk Laki- Laki WNI per Kelurahan di Jakarta Pusat')
# plt.savefig('kelurahan_wrong.png')
```

```python
# Membuat dataframe baru dengan groupby nama kecamatan dan menghitung jumlah laki - laki
kec_laki = table.groupby('NAMA KECAMATAN')['LAKI-LAKI WNI'].sum().reset_index()
```

```python
# Plot jumlah laki laki wni berdasarkan kecamatan
plt.figure(figsize=(15, 10))
x_label = kec_laki['NAMA KECAMATAN']
plt.bar(x=np.arange(len(x_label)),height=kec_laki['LAKI-LAKI WNI'])
plt.xticks(np.arange(len(x_label)), kec_laki['NAMA KECAMATAN'], rotation=30)
plt.xlabel('Kecamatan di Jakarta pusat')
plt.ylabel('Jumlah Penduduk Laki - Laki')
plt.title('Persebaran Jumlah Penduduk Laki- Laki WNI per Kecamatan di Jakarta Pusat')
plt.savefig('kecamatan.png')
```
